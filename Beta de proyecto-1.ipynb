## Configurar Almacenamiento de Datos con CSV

### Subtask:
Implementar las funciones para cargar y guardar los datos del inventario de forma persistente en un archivo CSV, incluyendo manejo de errores y validación de datos para asegurar la integridad.


**Reasoning**:
The first step is to define the data structure for the inventory items, including the keys (headers) that will be used in the CSV file and the in-memory representation. This prepares for the functions to load and save data.



```python
import csv
import os

# Define the headers for the inventory items
INVENTORY_HEADERS = [
    'SKU',
    'Nombre',
    'Descripción',
    'Precio de compra',
    'Precio de venta',
    'Categoría',
    'Proveedor',
    'Cantidad en stock',
    'Ubicación'
]

print("Inventory headers defined.")
```

**Reasoning**:
Now, I will implement the `load_inventory_from_csv` function. This function will read the CSV file, handle `FileNotFoundError`, convert numerical fields to appropriate data types, and return the inventory as a list of dictionaries.



```python
def load_inventory_from_csv(filepath):
    """
    Loads inventory data from a CSV file.

    Args:
        filepath (str): The path to the CSV file.

    Returns:
        list: A list of dictionaries, where each dictionary represents an inventory item.
              Returns an empty list if the file does not exist or is empty after headers.
    """
    inventory_data = []
    if not os.path.exists(filepath):
        print(f"File '{filepath}' not found. Creating an empty file with headers.")
        try:
            with open(filepath, 'w', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow(INVENTORY_HEADERS)
            return []
        except IOError as e:
            print(f"Error creating file '{filepath}': {e}")
            return []

    try:
        with open(filepath, 'r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            if not reader.fieldnames or reader.fieldnames != INVENTORY_HEADERS:
                print(f"Warning: CSV header mismatch or missing headers in '{filepath}'. Expected: {INVENTORY_HEADERS}, Got: {reader.fieldnames}")
                # If headers are missing or mismatched, we might want to handle this more robustly
                # For now, we'll try to proceed but this might lead to issues.
                # A more robust solution might involve creating a backup and re-initializing the file.

            for row in reader:
                processed_row = {}
                for header in INVENTORY_HEADERS:
                    value = row.get(header, '') # Get value, default to empty string if header missing
                    if header in ['Precio de compra', 'Precio de venta']:
                        try:
                            processed_row[header] = float(value) if value else 0.0
                        except ValueError:
                            print(f"Warning: Invalid float value for '{header}' in row: {row}. Defaulting to 0.0")
                            processed_row[header] = 0.0
                    elif header == 'Cantidad en stock':
                        try:
                            processed_row[header] = int(value) if value else 0
                        except ValueError:
                            print(f"Warning: Invalid integer value for '{header}' in row: {row}. Defaulting to 0")
                            processed_row[header] = 0
                    else:
                        processed_row[header] = value
                inventory_data.append(processed_row)
    except FileNotFoundError:
        # This case should be handled by the initial os.path.exists check, but kept for robustness
        print(f"Error: File '{filepath}' not found during read. Returning empty inventory.")
        return []
    except Exception as e:
        print(f"An unexpected error occurred while loading inventory from '{filepath}': {e}")
        return []
    return inventory_data

print("load_inventory_from_csv function defined.")
```

**Reasoning**:
Now I will implement the `save_inventory_to_csv` function. This function will write the current inventory data (list of dictionaries) back to the specified CSV file, ensuring that the headers are correctly written and handling potential file writing errors.



```python
def save_inventory_to_csv(filepath, inventory_data):
    """
    Saves inventory data to a CSV file.

    Args:
        filepath (str): The path to the CSV file.
        inventory_data (list): A list of dictionaries, where each dictionary represents an inventory item.
    """
    try:
        with open(filepath, 'w', newline='', encoding='utf-8') as file:
            writer = csv.DictWriter(file, fieldnames=INVENTORY_HEADERS)
            writer.writeheader()
            writer.writerows(inventory_data)
        print(f"Inventory successfully saved to '{filepath}'.")
    except (IOError, PermissionError) as e:
        print(f"Error saving inventory to '{filepath}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred while saving inventory to '{filepath}': {e}")

print("save_inventory_to_csv function defined.")
```

**Reasoning**:
The last instruction of the subtask is to initialize a global variable to hold the current inventory data, using the `load_inventory_from_csv` function. This step will create an `inventory` variable and populate it.



```python
INVENTORY_FILEPATH = 'inventory.csv'

inventory = load_inventory_from_csv(INVENTORY_FILEPATH)
print(f"Inventory initialized with {len(inventory)} items from '{INVENTORY_FILEPATH}'.")
```

## Desarrollar Lógica de Negocio del Inventario

### Subtask:
Implementar las funciones principales para gestionar el inventario, incluyendo añadir nuevos productos, eliminar productos existentes y actualizar las cantidades de stock.


**Reasoning**:
First, I will implement the `add_product` function. This function will handle adding new products to the inventory, including validation for unique SKUs, positive numerical values for price and quantity, and persistence to the CSV file.



```python
def add_product(sku, name, description, purchase_price, sales_price, category, supplier, initial_stock, location):
    """
    Adds a new product to the inventory.

    Args:
        sku (str): Unique product identifier.
        name (str): Product name.
        description (str): Product description.
        purchase_price (float): Price at which the product is purchased.
        sales_price (float): Price at which the product is sold.
        category (str): Product category.
        supplier (str): Product supplier.
        initial_stock (int): Initial quantity of the product in stock.
        location (str): Storage location of the product.

    Returns:
        bool: True if the product was added successfully, False otherwise.
    """
    global inventory

    # 1. Validate SKU uniqueness
    if any(item['SKU'] == sku for item in inventory):
        print(f"Error: SKU '{sku}' already exists. Product not added.")
        return False

    # 2. Validate positive numerical values
    if not isinstance(purchase_price, (int, float)) or purchase_price <= 0:
        print(f"Error: 'Precio de compra' must be a positive number. Got {purchase_price}.")
        return False
    if not isinstance(sales_price, (int, float)) or sales_price <= 0:
        print(f"Error: 'Precio de venta' must be a positive number. Got {sales_price}.")
        return False
    if not isinstance(initial_stock, int) or initial_stock < 0:
        print(f"Error: 'Cantidad inicial en stock' must be a non-negative integer. Got {initial_stock}.")
        return False

    # 3. Create new product dictionary and append to inventory
    new_product = {
        'SKU': sku,
        'Nombre': name,
        'Descripción': description,
        'Precio de compra': float(purchase_price),
        'Precio de venta': float(sales_price),
        'Categoría': category,
        'Proveedor': supplier,
        'Cantidad en stock': int(initial_stock),
        'Ubicación': location
    }
    inventory.append(new_product)

    # 4. Persist updated inventory
    save_inventory_to_csv(INVENTORY_FILEPATH, inventory)

    print(f"Success: Product '{name}' (SKU: {sku}) added to inventory.")
    return True

print("add_product function defined.")
```

**Reasoning**:
Now, I will implement the `delete_product` function. This function will locate a product by its SKU, remove it from the inventory if found, and then save the updated inventory to the CSV file.



```python
def delete_product(sku):
    """
    Deletes a product from the inventory based on its SKU.

    Args:
        sku (str): The SKU of the product to delete.

    Returns:
        bool: True if the product was deleted successfully, False otherwise.
    """
    global inventory

    initial_len = len(inventory)
    inventory = [item for item in inventory if item['SKU'] != sku]

    if len(inventory) < initial_len:
        save_inventory_to_csv(INVENTORY_FILEPATH, inventory)
        print(f"Success: Product with SKU '{sku}' deleted from inventory.")
        return True
    else:
        print(f"Error: Product with SKU '{sku}' not found. No product deleted.")
        return False

print("delete_product function defined.")
```

**Reasoning**:
Next, I will implement the `update_stock` function as per the subtask instructions. This function will modify the stock quantity of an existing product, validate the new stock level, and persist the changes.



```python
def update_stock(sku, quantity_change, reason=""):
    """
    Updates the stock quantity of a product based on its SKU.

    Args:
        sku (str): The SKU of the product to update.
        quantity_change (int): The amount to change the stock by (positive for additions, negative for reductions).
        reason (str, optional): The reason for the stock update (e.g., 'Purchase', 'Sale', 'Adjustment'). Defaults to "".

    Returns:
        bool: True if the stock was updated successfully, False otherwise.
    """
    global inventory

    found = False
    for item in inventory:
        if item['SKU'] == sku:
            found = True
            current_stock = item['Cantidad en stock']
            new_stock = current_stock + quantity_change

            # Validate that new_stock does not fall below zero
            if new_stock < 0:
                print(f"Error: Cannot update stock for SKU '{sku}'. New stock would be {new_stock}, which is below zero. Operation cancelled.")
                return False

            item['Cantidad en stock'] = new_stock
            save_inventory_to_csv(INVENTORY_FILEPATH, inventory)
            reason_msg = f" (Reason: {reason})" if reason else ""
            print(f"Success: Stock for product SKU '{sku}' updated from {current_stock} to {new_stock}{reason_msg}.")
            return True

    if not found:
        print(f"Error: Product with SKU '{sku}' not found. Stock not updated.")
        return False

print("update_stock function defined.")
```

## Implementar Generación de Reportes

### Subtask:
Desarrollar funciones para generar los reportes de 'Lista Completa de Productos', 'Productos con Bajo Stock' y 'Valor del Inventario' a partir de los datos existentes.


**Reasoning**:
First, I will implement the `generate_full_product_list_report` function. This function will return a copy of the entire `inventory` list, fulfilling the first instruction of the subtask.



```python
def generate_full_product_list_report():
    """
    Generates a report of all products in the inventory.

    Returns:
        list: A list of dictionaries, where each dictionary represents an inventory item.
    """
    global inventory
    print("Generating full product list report.")
    return list(inventory)

print("generate_full_product_list_report function defined.")
```

**Reasoning**:
Next, I will implement the `generate_low_stock_report` function. This function will take a threshold as input and return a list of products whose 'Cantidad en stock' is less than or equal to that threshold, fulfilling the second instruction of the subtask.



```python
def generate_low_stock_report(threshold):
    """
    Generates a report of products with stock quantity below or equal to a specified threshold.

    Args:
        threshold (int): The stock quantity threshold.

    Returns:
        list: A list of dictionaries, where each dictionary represents a product with low stock.
    """
    global inventory
    low_stock_products = [product for product in inventory if product.get('Cantidad en stock', 0) <= threshold]
    print(f"Generating low stock report for products with stock <= {threshold}.")
    return low_stock_products

print("generate_low_stock_report function defined.")
```

**Reasoning**:
Finally, I will implement the `generate_inventory_value_report` function. This function will calculate the total value of the inventory, allowing the user to specify whether to use 'Precio de compra' (purchase price) or 'Precio de venta' (sales price), thus completing the third instruction of the subtask.



```python
def generate_inventory_value_report(price_type='Precio de compra'):
    """
    Generates a report of the total inventory value based on a specified price type.

    Args:
        price_type (str): The type of price to use for calculation ('Precio de compra' or 'Precio de venta').
                          Defaults to 'Precio de compra'.

    Returns:
        float: The total calculated value of the inventory.
    """
    global inventory
    total_value = 0.0
    valid_price_types = ['Precio de compra', 'Precio de venta']

    if price_type not in valid_price_types:
        print(f"Error: Invalid price_type '{price_type}'. Must be one of {valid_price_types}.")
        return 0.0

    for product in inventory:
        stock = product.get('Cantidad en stock', 0)
        price = product.get(price_type, 0.0)
        total_value += stock * price

    print(f"Generating inventory value report using '{price_type}'. Total value: {total_value:.2f}.")
    return total_value

print("generate_inventory_value_report function defined.")
```
